### Биномиальное распределение 

```python
def binom_distrib(n,p,alpha = 0.05):
    q = 1-p #обр вероят
    m = n * p # мат ожид
    var = n*p*q # дисперсия бин распр 
    z_stat = stats.norm.ppf(1 - alpha / 2)
    C_l = int(round(m - z_stat * np.sqrt(var)))
    C_u = int(round(m + z_stat * np.sqrt(var)))
    return (C_l, C_u)
    
```

```python
# сортируем, тогда середина будет медианой 
alpha = 0.05
sort_lim_bal_0 = np.array(np.sort(data_0['LIMIT_BAL'].values))
sort_lim_bal_1 = np.array(np.sort(data_1['LIMIT_BAL'].values))

C_l_0, C_u_0 = binom_distrib(len(sort_lim_bal_0), 0.5)
C_l_1, C_u_1 = binom_distrib(len(sort_lim_bal_1), 0.5)
```

### Перестановочный критерий

Проверим гипотезу о равенстве распределений с помощью непараметрического перестановочного критерия

Идея заключается в том, чтобы найти такую группу перестановок исходной выборки X, что распределение X при справедливости нулевой гипотезы не отличается от распределения перестановочной выборки X.

$H_0\colon F_{X_1}(x) = F_{X_2}(x)$ (распределения выборок полностью равны)

$H_1\colon F_{X_1}(x) = F_{X_2}(x + \Delta), \Delta\neq 0$ (распределения выборок отличаются на величину сдвига)

```python
#Разность средних
def permutation_t_stat_ind(sample1, sample2):
    return np.mean(sample1) - np.mean(sample2)
```

```python
#Перестановки
def get_random_combinations(n1, n2, max_combinations):
    index = list(range(n1 + n2))
    indices = set([tuple(index)])
    for i in list(range(max_combinations - 1)):
        np.random.shuffle(index)
        indices.add(tuple(index))
    return [(index[:n1], index[n1:]) for index in indices]
```

```python
#Нулевое распределение 
def permutation_zero_dist_ind(sample1, sample2, max_combinations=None):
    joined_sample = np.hstack((sample1, sample2))
    n1 = len(sample1)
    n = len(joined_sample)
    
    if max_combinations:
        indices = get_random_combinations(n1, len(sample2), max_combinations)
    else:
        indices = [(list(index), filter(lambda i: i not in index, list(range(n)))) \
                    for index in itertools.combinations(list(range(n)), n1)]
    
    distr = [joined_sample[list(i[0])].mean() - joined_sample[list(i[1])].mean() \
             for i in indices]
    return distr
```

```python
zero_distrib = permutation_zero_dist_ind(sort_lim_bal_0, sort_lim_bal_1,max_combinations=1000)

plt.figure(figsize=(12,8))
g1 = sns.histplot(zero_distrib,palette=['#FF0000'],alpha=0.9)
#g1.set_xlim([0, 10000])

plt.title('Нулевое распределение',fontsize=18)
plt.xlabel('Delta_mean',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.show()
```

```python
#Перестановочный тест
def permutation_test(sample1, sample2, max_permutations=None, alternative='two-sided'):
    if alternative not in ('two-sided', 'less', 'greater'):
        raise ValueError("alternative not recognized\n"
                         "should be 'two-sided', 'less' or 'greater'")
    
    t_stat = permutation_t_stat_ind(sample1, sample2)
    
    zero_distr = permutation_zero_dist_ind(sample1, sample2, max_permutations)
    
    if alternative == 'two-sided':
        return sum([1. if abs(x) >= abs(t_stat) else 0. for x in zero_distr]) / len(zero_distr)
    
    if alternative == 'less':
        return sum([1. if x <= t_stat else 0. for x in zero_distr]) / len(zero_distr)

    if alternative == 'greater':
        return sum([1. if x >= t_stat else 0. for x in zero_distr]) / len(zero_distr)
```

```python
print('p-value:', permutation_test(sort_lim_bal_0, sort_lim_bal_1, max_permutations=1000))
```

p-value < 0.05 следовательно нулевая гипотеза о равенстве распределений отвергается. Значит, распределения выборок отличаются на величину сдвига, тогда для группы "вернувших" кредитный лимит выше.

### Доверительный интервал для разности двух долей независимых выборок

$\Large\widehat{p_1} - \widehat{p_2} - z_{\alpha/2} * \sqrt{\frac{p_1 * q_1}{n_1}+\frac{p_2 * q_2}{n_2}} < p_1-p_2 <\widehat{p_1} - \widehat{p_2} + z_{\alpha/2} * \sqrt{\frac{p_1 * q_1}{n_1}+\frac{p_2 * q_2}{n_2}}$

```python
def proportions_confint_diff_ind(sample1, sample2, alpha = 0.05):    
    z = stats.norm.ppf(1 - alpha / 2.)   
    p1 = float(sum(sample1)) / len(sample1)
    p2 = float(sum(sample2)) / len(sample2)
    
    С_left = (p1 - p2) - z * np.sqrt(p1 * (1 - p1) / len(sample1) + p2 * (1 - p2) / len(sample2))
    С_right= (p1 - p2) + z * np.sqrt(p1 * (1 - p1) / len(sample1) + p2 * (1 - p2) / len(sample2))
    
    return (С_left, С_right)
```

```python
proportions_confint_diff_ind(abs(data_0['SEX'].values-2),abs(data_1['SEX'].values-2)) # 0-жен, 1-муж
```

В доверительный интервал не входит 0 и он находится слева от него, соответственно с 95%-ной уверенностью мы можем утверждать, что доля мужчин не вернувших кредит больше, чем доля вернувших.

### Z-критерий для доли для двух независимых выборок

$H_0: p_0 = p_1$ доля выплативших мужчин не отличается от доли не выплативших.  
$H_1: p_0 \neq p_1$ доля выплативших мужчин отличается от доли не выплативших. 

```python
def proportions_diff_z_stat_ind(sample1, sample2):
    n1 = len(sample1)
    n2 = len(sample2)
    
    p1 = float(sum(sample1)) / n1
    p2 = float(sum(sample2)) / n2 
    P = float(p1*n1 + p2*n2) / (n1 + n2)
    
    return (p1 - p2) / np.sqrt(P * (1 - P) * (1. / n1 + 1. / n2))
```

```python
def proportions_diff_z_test(z_stat, alternative='two-sided'):
    if alternative not in ('two-sided', 'less', 'greater'):
        raise ValueError("alternative not recognized\n"
                         "should be 'two-sided', 'less' or 'greater'")
    
    if alternative == 'two-sided':
        return 2 * (1 - stats.norm.cdf(np.abs(z_stat)))
    
    if alternative == 'less':
        return stats.norm.cdf(z_stat)

    if alternative == 'greater':
        return 1 - stats.norm.cdf(z_stat)
```

```python
z_test = proportions_diff_z_test(proportions_diff_z_stat_ind(abs(data_0['SEX'].values-2),abs(data_1['SEX'].values-2)))
print('p-value:',"{:.12f}".format(float(z_test)) )
```

p-value < 0.05 следовательно нулевая гипотеза о равенстве доли выплативших и не выплативших мужчин отличается. Значит гендерный состав отличается

# Семейное положение (MARRIAGE):

# 4.Значение V Крамера

Измерим силу связи между двумя категориальными признаками: "MARRIAGE", "default" при помощи значения V Крамера.

```python
marriage_piv_table = pd.pivot_table(
    data[["MARRIAGE", "default", "LIMIT_BAL"]],
    values=["LIMIT_BAL"],
    index=["MARRIAGE"],
    columns=["default"],
    fill_value = 0,
    aggfunc='count')

marriage_piv_table.columns = ['Вернул', 'Не вернул'] 
marriage_piv_table.index = ['Отказываюсь отвечать','Замужем/Женат','Холост','Нет данных']
```

```python
marriage_piv_table
```

```python
def cramers_stat(matrix):
    chi2 = stats.chi2_contingency(matrix)[0]
    n = matrix.sum()
    return np.sqrt(chi2 / (n*(min(matrix.shape)-1)))
```

```python
cramers_stat(marriage_piv_table.values)
```

```python
print('Значение V Крамера:',cramers_stat(marriage_piv_table.values))
```

Значение V Крамера достаточно низкое, следовательно семейный статус практически не влияет на возврат долга.
